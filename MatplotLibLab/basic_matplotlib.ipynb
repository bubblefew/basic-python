```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
y = np.array([1050, 1100, 1150, 510, 1250, 500, 1350, 100])
```

```python
plt.plot(x, y)
plt.xticks(x) # ปัดเลขให้เป็นจำนวนเต็ม
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

### Bar Chart

```python
plt.bar(x, y)
plt.xticks(x) # ปัดเลขให้เป็นจำนวนเต็ม
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

### csv to chart

```python
df = pd.read_csv("datasets/gdp.csv")
df

# create a chart with GDP data
plt.plot(df['Year'], df['GDP'])
plt.xticks(df['Year'])
plt.xlabel('Year')
plt.title('GDP of Thailand')
plt.ylabel('GDP (Billion Baht)')
plt.show()
```

```python
plt.bar(
    df['Year'],
    df['GDP'],
    color=['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray'],
    edgecolor='black'
)
plt.xticks(df['Year'])
plt.xlabel('Year')
plt.title('GDP of Thailand')
plt.ylabel('GDP (Billion Baht)')
plt.show()
```

```python
plt.pie(
    df['GDP'],
    labels=df['Year'],
    autopct='%.0f%%',
    colors=['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray']
)
plt.title('GDP of Thailand')
plt.show()
```

# Scatter plot

```python
plt.scatter(df['Year'], df['GDP'])
plt.xticks(df['Year'])
plt.xlabel('Year')
plt.title('GDP of Thailand')
plt.ylabel('GDP (Billion Baht)')
plt.show()
```
